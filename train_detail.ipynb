```python
import torch
from utils.utils import *
from utils.model import DPSC
import matplotlib.pyplot as plt
import yaml
```

```python
data_path = './diff_data/mnist_diff_ft.pt'
data, label = loaddata(data_path)
```

```python
with open('./config.yaml') as config_file:
    config_all = yaml.load(config_file, yaml.Loader)
config = config_all['mnist']
```

```python
net = DPSC(n_all=len(label), label=label, alpha=config['alpha'], gamma1=config['gamma1'],
               gamma2=config['gamma2'], dim=config['dim'], ro=config['ro'])
loss, nmis, accs = net.train(data=data, epochs=config['epoch'], ft=True)
```

```python
x = np.linspace(20, config['epoch'], int(config['epoch']/20))
plt.figure(dpi=300)
plt.plot(np.linspace(1, config['epoch'], config['epoch']).astype(int),scale_loss(loss),label='Loss')
plt.plot(x,accs,label='ACC')
plt.scatter(x,accs,marker='o')
plt.plot(x,nmis,label='NMI')
plt.scatter(x,nmis,marker='x')
plt.xlabel('epochs')
plt.legend()
plt.show()
```
